```python
# dropped all duplicated rows
df_no_dup=df.drop_duplicates()
df_no_dup.duplicated().sum()
```

```python
df_divided_genres=df_no_dup.copy()
```

```python
df_no_dup[['lead_genre','genre_2','genre_3','genre_4','genre_5','genre_6','genre_7','genre_8','genre_9','genre_10']]=df_no_dup['genres'].str.split(',', expand=True)
df_no_dup
```

```python
genres=['lead_genre','genre_2','genre_3','genre_4','genre_5','genre_6','genre_7','genre_8','genre_9','genre_10']
for col in genres:
    df_no_dup[col]=df_no_dup[col].str.strip(" ''['']")
df_no_dup
```

```python
df_no_dup.info()
```

```python
# most genre columns were empty so we decided to frop them and focus on the rest
df_no_dup=df_no_dup.drop(['genres','genre_3','genre_4','genre_5','genre_6','genre_7','genre_8','genre_9','genre_10'],axis=1)
df_no_dup
```

```python
df_no_dup.to_csv("game_score_database_no_dup_expanded_genres.csv",index=False)
```

```python
# we decided to fill the empty user score with the median
df_fill_na=df_no_dup.copy()
df_fill_na["user_score"].fillna(df_fill_na.user_score.median(),inplace=True)
df_fill_na.info()
```

```python
# most games are rated T for teen so we decided to fill missing values with it
df_fill_na["rating"].fillna(df_fill_na.rating.mode()[0],inplace=True)
```

```python
# all other missing values were about 100 so we decided to drop them all
df_fill_na.dropna(inplace=True)
df_fill_na.info()
```

```python
# from our personal understanding of video games we noticed that most of the genres weren't indicative of their true genre
# as in this example
df_fill_na[df_fill_na["lead_genre"]=='Sim']
```

```python
df_fill_na.lead_genre.unique()
```

```python
df_fill_na[df_fill_na["lead_genre"]=='First-Person']
```

```python
# we noticed that genre 2 column was more reflective of the true genres and decided to drop the lead genre column
df_fill_na.genre_2.unique()
```

```python
#using raplace method with regex option to lower the dimension of the genre column, every genre is replaced with a corresponding bigger genre
df_fill_na['genre_2']=df_fill_na['genre_2'].str.replace(r'(Fantasy|General|Visual Novel|Adventure|Third-Person|3D|Sci-Fi|2D|Text|Point-and-Click|Trainer|Sandbox|Hidden Object|Videos|Platformer)', 'adventure', regex=True)
df_fill_na['genre_2']=df_fill_na['genre_2'].str.replace(r'(Action RPG|First-Person|PC-style RPG|Japanese-Style|Console-style RPG|Role-Playing|Western-Style|Marine|Massively Multiplayer|Massively Multiplayer Online|MOBA|RPG Online)', 'RPG', regex=True)
df_fill_na['genre_2']=df_fill_na['genre_2'].str.replace(r'(Turn-Based|Real-Time|Wargame|Business / Tycoon|Tactical)', 'Strategy', regex=True)
df_fill_na['genre_2']=df_fill_na['genre_2'].str.replace(r'(Alternative|Traditional|Team|Car Combat|Individual|Sports|Exercise / Fitness|Football|Nature|Soccer|Olympic Sports|Other|Gambling|Biking|Volleyball|Racing|Simulation|GT / Street|Driving|Demolition Derby|Rally / Offroad)', 'sports', regex=True)
df_fill_na['genre_2']=df_fill_na['genre_2'].str.replace(r'(Flight|Virtual|Management|Tycoon|Small Spaceship|Virtual Life|Breeding/Constructing|City Building|Train|Application)', 'simulator', regex=True)
df_fill_na['genre_2']=df_fill_na['genre_2'].str.replace(r'(Modern|Open-World|Action adventure|Shooter|Horror|Compilation|Linear|Survival|"Beat-\'Em-Up"|Action|Historic|Scrolling|Submarine|Mission-based|Tank|Roguelike|Large Spaceship|Space|Vehicle|Ship|Rail|Fighting)', 'action', regex=True)
df_fill_na['genre_2']=df_fill_na['genre_2'].str.replace(r'(Stacking|Puzzle|Rhythm|Kart|Futuristic|Card Battle|Logic|Pinball|Arcade|Matching|Board / Card Game|Board Games|Music Maker|Trivia / Game Show|Edutainment|Party / Minigame|Parlor|Music|"Shoot-\'Em-Up"|Miscellaneous|Party)', 'arcade', regex=True)
```

```python
#dealing with special cases
df_fill_na.replace('Sim','sports',inplace=True)
df_fill_na.replace('simulator Life','simulator',inplace=True)
df_fill_na.replace('RPG Online', 'RPG',inplace=True)
```

```python
df_fill_na.genre_2.value_counts()
```

```python
df_fill_na=df_fill_na.drop(['lead_genre'],axis=1)
df_fill_na
```

```python
# rearranging the columns to put the score (destination column) in the end
df_fill_na=df_fill_na[["game_name","user_score","platform","developer","developer_score","publisher","publisher_score","rating","release_month","release_year","num_crit_review","num_pos_critic","num_mix_critic","num_neg_critic","descriptions","genre_2","score"]]
df_fill_na.to_csv("dataframe_for_visual.csv",index=False)
df_fill_na
```
