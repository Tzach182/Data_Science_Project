```python
def get_column_ohe(d, feature)    
    
    # Get one hot encoding of columns ...
    one_hot = pd.get_dummies(d[f'{feature}'])
    # Drop column B as it is now encoded
    d = d.drop(f'{feature}',axis = 1)
    # Join the encoded df
    d = d.join(one_hot)
    
    return d

list_of_cat = ["a", "b", "c"]

for k in list_of_cat:
    d = get_column_ohe(d, k)
```
