```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("dataframe_for_visual.csv")
df
```

```python
#https://www.kaggle.com/code/garfieldliang/video-games-analysis
#distribution of games across platforms
platGenre = pd.crosstab(df.platform,df.genre_2)
platGenreTotal = platGenre.sum(axis=1).sort_values(ascending = False)
plt.figure(figsize=(8,6))
sns.barplot(y = platGenreTotal.index, x = platGenreTotal.values, orient='h')
plt.ylabel = "Platform"
plt.xlabel = "The amount of games"
plt.show()
```

```python
platGenre['Total'] = platGenre.sum(axis=1)
popPlatform = platGenre[platGenre['Total']>1000].sort_values(by='Total', ascending = False)
neededdata = popPlatform.loc[:,:'Strategy']
maxi = neededdata.values.max()
#maxi = popPlatform.values.max()
#mini = popPlatform.values.min()
mini = neededdata.values.min()
popPlatformfinal = popPlatform.append(pd.DataFrame(popPlatform.sum(), columns=['total']).T, ignore_index=False)
sns.set(font_scale=0.7)
plt.figure(figsize=(15,10))
sns.heatmap(popPlatformfinal, vmin = mini, vmax = maxi, annot=True, fmt="d")
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.countplot(x="genre_2", data=df, order = df['genre_2'].value_counts().index)
plt.xticks(rotation=90)
```

```python
#https://www.kaggle.com/code/snanilim/video-games-sales-analysis-and-visualization
plt.figure(figsize=(15, 10))
sns.countplot(x="release_year", data=df, order = df.groupby(by=['release_year'])['game_name'].count().sort_values(ascending=False).index)
plt.xticks(rotation=90)
```

```python
data_year = df.groupby(by=['release_year'])['score'].mean()
data_year = data_year.reset_index()
plt.figure(figsize=(15, 10))
sns.barplot(x="release_year", y="score", data=data_year)
plt.xticks(rotation=90)
```

```python
year_mean_score_dx = df.groupby(by=['release_year', 'genre_2'])['score'].mean().reset_index()
year_score = year_mean_score_dx.groupby(by=['release_year'])['score'].transform(max) == year_mean_score_dx['score']
year_score_max = year_mean_score_dx[year_score].reset_index(drop=True)
# year_sale_max
genre = year_score_max['genre_2']
```

```python
plt.figure(figsize=(30, 18))
g = sns.barplot(x='release_year', y='score', data=year_score_max)
index = 0
for value in year_score_max['score']:
    g.text(index, value + 1, str(genre[index] + '----' +str(round(value, 2))), color='#000', size=14, rotation= 90, ha="center")
    index += 1

plt.xticks(rotation=90)
plt.show()
```

```python
data_platform = df.groupby(by=['platform'])['score'].mean()
data_platform = data_platform.reset_index()
data_platform = data_platform.sort_values(by=['score'], ascending=False)
# data_platform
```

```python
plt.figure(figsize=(15, 10))
sns.barplot(x="platform", y="score", data=data_platform)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(13,10))
sns.heatmap(df.corr(), cmap = "Blues", annot=True, linewidth=3)
```

```python
#https://www.kaggle.com/code/mayurdalvi/video-game-data-visualization
df['user_score'] = df['user_score'].astype(np.float)
top_user_score = df.sort_values(['user_score'], ascending=False)[:10]
top_user_score[['game_name','user_score']]
```

```python
fig, ax = plt.subplots(figsize = (10,5))
sns.barplot(x = top_user_score['user_score'], y = top_user_score['game_name'])
plt.xlim(9.3,9.8)
plt.title('Top 10 Games by User Review ', fontsize = 16)
plt.xlabel('User Review Score', fontsize = 14)
plt.ylabel('Name', fontsize = 14)
```

```python
top_meta_score = df.sort_values(['score'], ascending = False)[:10]
top_meta_score[['game_name','score']]
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=top_meta_score['score'], y=top_meta_score['game_name'], orient = 'h', palette='flare')
plt.xlim(95,100)
plt.title('Top 10 Games by Meta Score', fontsize=16)
plt.xlabel('score', fontsize=14)
plt.ylabel('game_Name', fontsize=14)
```
